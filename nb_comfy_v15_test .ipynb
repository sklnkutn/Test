```python
# ЯЧЕЙКА 1: только проверка интернет-канала (не блокирует работу ноутбука).
import os  # Импорт os нужен для чтения переменных окружения (если пользователь захочет переопределить пороги).
import re  # Импорт re нужен для извлечения среднего ping из текстового вывода команды ping.
import subprocess  # Импорт subprocess нужен для запуска системной команды ping.
import sys  # Импорт sys нужен для обращения к текущему интерпретатору Python (sys.executable).

MIN_DOWNLOAD_MBPS = float(os.environ.get("MIN_DOWNLOAD_MBPS", "300"))  # Константа задаёт минимально желаемую скорость скачивания в Мбит/с.
MAX_PING_MS = float(os.environ.get("MAX_PING_MS", "100"))  # Константа задаёт максимально желаемый средний ping в миллисекундах.
PING_HOSTS = ["civitai.com", "huggingface.co"]  # Список сайтов для проверки ping (критерий ping остаётся прежним).


def ensure_speedtest_package():
    """Функция проверяет наличие speedtest и устанавливает пакет, если он отсутствует."""
    import importlib.util as importlib_util  # Локальный импорт для проверки существования Python-модуля.
    module_exists = importlib_util.find_spec("speedtest") is not None  # Флаг показывает, установлен ли модуль speedtest.
    if module_exists:
        print("speedtest-cli: уже установлен")
        return True
    print("speedtest-cli: не найден, устанавливаю...")
    install_proc = subprocess.run(  # Результат выполнения pip install сохраняется в переменную install_proc.
        [sys.executable, "-m", "pip", "install", "-U", "speedtest-cli"],
        text=True,
        capture_output=True,
    )
    if install_proc.returncode != 0:  # Если код возврата не нулевой, установка не удалась.
        print("[WARN] Не удалось установить speedtest-cli. Проверка скорости будет пропущена.")
        print(install_proc.stdout[-800:])
        print(install_proc.stderr[-800:])
        return False
    return True


def run_speedtest():
    """Функция измеряет download/upload скорость и возвращает словарь результатов."""
    import speedtest  # Импорт внутри функции, чтобы не ломать весь ноутбук, если пакет не установлен.
    result = {"download_mbps": None, "upload_mbps": None, "error": None}  # Словарь result хранит результат теста сети.
    try:
        st = speedtest.Speedtest(secure=True)  # Объект st выполняет подбор сервера и сами измерения.
        st.get_best_server()
        download_bps = st.download()  # Скорость скачивания в битах в секунду.
        upload_bps = st.upload()  # Скорость отдачи в битах в секунду.
        result["download_mbps"] = download_bps / 1_000_000  # Перевод в Мбит/с.
        result["upload_mbps"] = upload_bps / 1_000_000  # Перевод в Мбит/с.
    except Exception as exc:
        result["error"] = str(exc)  # В поле error сохраняется причина сбоя speedtest.
    return result


def avg_ping(host):
    """Функция запускает ping до host и возвращает среднее значение в мс."""
    cmd = ["ping", "-c", "3", "-W", "2", host]  # Команда ping: 3 пакета и таймаут 2 секунды на ответ.
    try:
        ping_proc = subprocess.run(cmd, text=True, capture_output=True)  # Результат ping сохраняем в ping_proc.
    except FileNotFoundError:
        return None, "Команда ping недоступна в этом окружении"
    if ping_proc.returncode != 0:
        return None, ping_proc.stderr.strip() or ping_proc.stdout.strip()  # Возвращаем ошибку, если ping неуспешен.
    summary_match = re.search(r"=\s*([\d\.]+)/([\d\.]+)/([\d\.]+)/", ping_proc.stdout)  # Поиск строки min/avg/max.
    if not summary_match:
        return None, "Не удалось распарсить средний ping"
    avg_ms = float(summary_match.group(2))  # Второе число в шаблоне — это средний ping (avg).
    return avg_ms, None


print("Проверка интернета (требования: download >= 300 Мбит/с и ping < 100 мс)...")
can_run_speedtest = ensure_speedtest_package()  # Флаг показывает, доступен ли speedtest для измерения скорости.
net = {"download_mbps": None, "upload_mbps": None, "pings": {}}  # Словарь net хранит итоговые сетевые метрики.

if can_run_speedtest:
    speed_result = run_speedtest()  # Переменная speed_result хранит результат speedtest.
    net["download_mbps"] = speed_result.get("download_mbps")
    net["upload_mbps"] = speed_result.get("upload_mbps")
    if speed_result.get("error"):
        print(f"[WARN] Ошибка speedtest: {speed_result['error']}")
else:
    print("[WARN] Speedtest пропущен (модуль недоступен).")

for host in PING_HOSTS:
    host_ping, host_err = avg_ping(host)  # host_ping — средний ping, host_err — текст ошибки.
    net["pings"][host] = {"avg_ms": host_ping, "error": host_err}

print("\n--- Результаты сети ---")
print(f"Download: {net['download_mbps']:.2f} Мбит/с" if net["download_mbps"] is not None else "Download: N/A")
print(f"Upload:   {net['upload_mbps']:.2f} Мбит/с" if net["upload_mbps"] is not None else "Upload:   N/A")
for host, host_data in net["pings"].items():
    print(f"Ping {host}: {host_data['avg_ms']:.2f} мс" if host_data["avg_ms"] is not None else f"Ping {host}: N/A ({host_data['error']})")

speed_ok = net["download_mbps"] is not None and net["download_mbps"] >= MIN_DOWNLOAD_MBPS  # Проверка нового критерия скорости (>=300).
ping_ok = all(host_data["avg_ms"] is not None and host_data["avg_ms"] < MAX_PING_MS for host_data in net["pings"].values())  # Проверка старого критерия ping.

if speed_ok and ping_ok:
    print("Сеть соответствует требованиям для загрузок и работы.")
else:
    print("Предупреждение: сеть ниже рекомендуемого уровня (download < 300 Мбит/с и/или ping >= 100 мс).")
```

```python
# ЯЧЕЙКА 3: корректный порядок — сначала ComfyUI, потом зависимости, потом создание остальных директорий.
import shutil  # Импорт shutil нужен для удаления конфликтной директории ComfyUI без создания backup-папок.
import subprocess  # Импорт subprocess нужен для запуска apt/pip/git команд.
import sys  # Импорт sys нужен для обращения к текущему интерпретатору Python (sys.executable).
from pathlib import Path  # Импорт Path нужен для проверки наличия папок/файлов.

COMFY_REPO_URL = "https://github.com/comfyanonymous/ComfyUI.git"  # URL официального репозитория ComfyUI.
APT_PACKAGES_AFTER_CLONE = ["python3-venv", "python3-pip", "aria2", "libgl1", "libglib2.0-0"]  # Системные пакеты, которые ставим после получения ComfyUI.
PYTHON_PACKAGES = ["tqdm", "huggingface_hub", "gdown"]  # Базовые Python-пакеты для последующих ячеек загрузки.
IPADAPTER_PLUS_PACKAGES = ["insightface", "onnxruntime", "opencv-python-headless"]  # Пакеты для корректной работы ComfyUI_IPAdapter_plus.


def run_cmd(cmd, check=True):
    """Функция запускает системную команду, печатает её и возвращает объект результата."""
    print("[CMD]", " ".join(str(x) for x in cmd))
    return subprocess.run(cmd, check=check, text=True, capture_output=True)


def run_cmd_with_logs(cmd, check=False):
    """Функция запускает команду и печатает последние строки stdout/stderr для диагностики."""
    result = run_cmd(cmd, check=check)
    if result.stdout:
        print(result.stdout[-3000:])
    if result.stderr:
        print(result.stderr[-3000:])
    return result


def ensure_comfy_repo():
    """Функция гарантирует наличие актуального git-репозитория ComfyUI в COMFY_DIR без backup-папок."""
    comfy_path = Path(COMFY_DIR)  # comfy_path — целевой путь репозитория ComfyUI.
    comfy_git_dir = comfy_path / ".git"  # comfy_git_dir — путь к .git внутри COMFY_DIR.

    if comfy_git_dir.exists():
        print("ComfyUI уже git-репозиторий, выполняю git pull...")
        run_cmd_with_logs(["git", "-C", str(comfy_path), "pull", "--ff-only"], check=False)
        return

    if comfy_path.exists() and any(comfy_path.iterdir()):
        print(f"Найдена непустая не-git папка: {comfy_path}")
        print("Удаляю её без создания backup, чтобы clone прошёл в чистую директорию.")
        shutil.rmtree(comfy_path)

    print("ComfyUI не найден, выполняю git clone...")
    clone_result = run_cmd_with_logs(["git", "clone", COMFY_REPO_URL, str(comfy_path)], check=False)
    if clone_result.returncode != 0:
        raise RuntimeError("Не удалось клонировать ComfyUI. Смотрите stderr выше (частая причина: сеть/доступ к GitHub).")


run_cmd_with_logs(["apt", "update", "-qq"], check=False)
run_cmd_with_logs(["apt", "install", "-y", "-qq", "git"], check=False)  # Ставим git до clone/pull.
ensure_comfy_repo()  # 1) Сначала гарантируем наличие ComfyUI.

run_cmd_with_logs(["apt", "install", "-y", "-qq", *APT_PACKAGES_AFTER_CLONE], check=False)  # 2) Ставим остальные системные зависимости.
run_cmd_with_logs([sys.executable, "-m", "pip", "install", "--upgrade", "pip", "setuptools", "wheel"], check=False)
run_cmd_with_logs([sys.executable, "-m", "pip", "install", *PYTHON_PACKAGES], check=False)  # Ставим базовые пакеты без форс-апдейта всего окружения.
run_cmd_with_logs([sys.executable, "-m", "pip", "install", *IPADAPTER_PLUS_PACKAGES], check=False)  # Ставим зависимости IPAdapter_plus.

requirements_file = Path(COMFY_DIR) / "requirements.txt"  # requirements_file — путь к requirements ComfyUI.
if requirements_file.exists():
    run_cmd_with_logs([sys.executable, "-m", "pip", "install", "-r", str(requirements_file)], check=False)

# 3) После clone/install создаём рабочие папки (кроме BASE_DIR и COMFY_DIR, они уже существуют/управляются отдельно).
for required_dir in [VOLUME_DIR, MODELS_DIR, CHECKPOINTS_DIR, LORA_DIR, UPSCALER_DIR, VAE_DIR, CONTROLNET_DIR, IPADAPTER_DIR, CUSTOM_NODES_DIR, CLIP_VISION_DIR, ULTRALYTICS_DIR, BBOX_DIR, SEGM_DIR, SAMS_DIR]:
    Path(required_dir).mkdir(parents=True, exist_ok=True)

print("Ячейка 3 завершена: ComfyUI, зависимости и директории готовы.")
```

```python
# ЯЧЕЙКА 4: формируем списки и выполняем параллельные загрузки (до 5 потоков, приоритет: тяжёлые -> лёгкие).
import os  # Импорт os нужен для чтения токенов и настроек из переменных окружения.
import subprocess  # Импорт subprocess нужен для git/aria2c/curl команд.
import time  # Импорт time нужен для пауз между повторными попытками скачивания.
from pathlib import Path  # Импорт Path нужен для работы с путями и создания папок.
from concurrent.futures import ThreadPoolExecutor, as_completed  # Импорт для параллельных скачиваний.

MAX_PARALLEL_DOWNLOADS = int(os.environ.get("MAX_PARALLEL_DOWNLOADS", "3"))  # Максимум параллельных загрузок (расширено до 5).
MIN_VALID_FILE_BYTES = int(os.environ.get("MIN_VALID_FILE_BYTES", "1024"))  # Минимальный валидный размер файла.
INVALID_TEXT_BYTES_LIMIT = int(os.environ.get("INVALID_TEXT_BYTES_LIMIT", str(5 * 1024 * 1024)))  # До какого размера файл проверяем как потенциально текстовую ошибку.
FORCE_REDOWNLOAD = os.environ.get("FORCE_REDOWNLOAD", "0").lower() in {"1", "true", "yes", "y"}  # Флаг принудительного перекачивания.
CIVITAI_TOKEN = os.environ.get("CIVITAI_TOKEN", "")  # Токен Civitai для приватных/ограниченных ссылок.
HF_TOKEN = os.environ.get("HF_TOKEN", "")  # Токен Hugging Face для gated/ограниченных ссылок.
ARIA2C_AVAILABLE = subprocess.run(["bash", "-lc", "command -v aria2c >/dev/null"], check=False).returncode == 0  # Проверяем наличие aria2c.
DOWNLOAD_RETRIES = int(os.environ.get("DOWNLOAD_RETRIES", "3"))  # Число повторных попыток скачивания каждого файла.
DOWNLOAD_RETRY_DELAY_SEC = int(os.environ.get("DOWNLOAD_RETRY_DELAY_SEC", "6"))  # Пауза между повторными попытками.
CURL_TIMEOUT_SEC = int(os.environ.get("CURL_TIMEOUT_SEC", "1800"))  # Таймаут одной попытки через curl.

CUSTOM_NODE_REPOS = [  # Список custom nodes, обязательных по ТЗ.
    "https://github.com/ltdrdata/ComfyUI-Manager",
    "https://github.com/cubiq/ComfyUI_IPAdapter_plus",
    "https://github.com/Fannovel16/comfyui_controlnet_aux",
    "https://github.com/ssitu/ComfyUI_UltimateSDUpscale",
    "https://github.com/ltdrdata/ComfyUI-Impact-Pack",
    "https://github.com/ltdrdata/ComfyUI-Impact-Subpack",
]
# Список URL базовых моделей (checkpoints). Можно указывать строку URL или кортеж (URL, "ИмяФайла.safetensors").
CHECKPOINT_URLS = [
    #WAI
    #("https://civitai.com/api/download/models/2514310?type=Model&format=SafeTensor&size=pruned&fp=fp16", "wai.safetensors"),
    #Juggernaut_XL
    #("https://civitai.com/api/download/models/1759168", "juggernaut_XL.safetensors"),
    #Pony checkpoints
    ("https://civitai.com/api/download/models/290640", "pony.safetensors"),
    #CyberRealistic Pony
    ("https://civitai.com/api/download/models/2581228?type=Model&format=SafeTensor&size=full&fp=fp32", "CyberRealisticPony.safetensors"),
    #Pony Realism
    ("https://civitai.com/api/download/models/914390?type=Model&format=SafeTensor&size=full&fp=fp16", "PonyRealism.safetensors"),
]
 # Список URL LoRA моделей.
LORA_URLS = [
    #Illustrious LoRA
    #("https://civitai.com/api/download/models/1736373?type=Model&format=SafeTensor", "detailer_IL_V2.safetensors"),
    #("https://civitai.com/api/download/models/1124771?type=Model&format=SafeTensor", "realistic_filter_V1.safetensors"),
    #("https://civitai.com/api/download/models/1914557?type=Model&format=SafeTensor", "hyperrealistic_v4_ill.safetensors"),
    #("https://civitai.com/api/download/models/1882710?type=Model&format=SafeTensor", "niji_semi_realism_v35_ill.safetensors"),
    #("https://civitai.com/api/download/models/1711464?type=Model&format=SafeTensor", "atnr_style_ill_v11.safetensors"),
    #("https://civitai.com/api/download/models/1839268?type=Model&format=SafeTensor", "face_enhancer_ill.safetensors"),
    #("https://civitai.com/api/download/models/2196453?type=Model&format=SafeTensor", "smooth_detailer_booster_v4.safetensors"),
    #("https://civitai.com/api/download/models/2555444?type=Model&format=SafeTensor", "usnr_style_v-pred.safetensors"),
    #("https://civitai.com/api/download/models/1056404?type=Model&format=SafeTensor", "748cm_style_v1.safetensors"),
    #("https://civitai.com/api/download/models/2620790?type=Model&format=SafeTensor", "velvets_mythic_fantasy_styles_il.safetensors"),
    #("https://civitai.com/api/download/models/2661972?type=Model&format=SafeTensor", "pixel_art_style_il_v7.safetensors"),
    #Pony Style LoRA
    ("https://civitai.com/api/download/models/1558543", "s_abstract_painting_style_pony_v6_xl.safetensors"),
    ("https://civitai.com/api/download/models/436219", "s_incase_style_v3_1_ponyxl_ilff.safetensors"),
    ("https://civitai.com/api/download/models/1091750", "s_chp_pony.safetensors"),
    ("https://civitai.com/api/download/models/1029540", "s_vixons_pony_gothic_styles.safetensors"),
    ("https://civitai.com/api/download/models/1169083", "s_realistic_animeixl_v2.safetensors"),
    ("https://civitai.com/api/download/models/522995", "s_popyays_epic_fantasy_style.safetensors"),
    ("https://civitai.com/api/download/models/1122823", "s_MoriiMeeGothicNiji.safetensors"),
    ("https://civitai.com/api/download/models/839103", "s_cutesexyrobutts.safetensors"),
    ("https://civitai.com/api/download/models/481798", "s_sinozick.safetensors"),
    ("https://civitai.com/api/download/models/373076", "s_vintage_1990.safetensors"),
    ("https://civitai.com/api/download/models/486749", "s_vixons_pony_styles.safetensors"),
    #Pony detail LoRA
    ("https://civitai.com/api/download/models/1242203", "d_dramatic_lighting.safetensors"),
    ("https://civitai.com/api/download/models/1506035", "d_add_detail.safetensors"),
    ("https://civitai.com/api/download/models/708275", "d_background_detail.safetensors"),
    ("https://civitai.com/api/download/models/596040", "d_real_beauty.safetensors"),
]
# Список URL upscaler моделей.
UPSCALER_URLS = [
    #WAI default Upscaler
    #("https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth", "RealESRGAN_x4plus.pth"),
    #Pony default Upscaler
    ("https://huggingface.co/Kim2091/UltraSharp/resolve/main/4x-UltraSharp.pth", "4x-UltraSharp.pth"),
    ("https://huggingface.co/FacehugmanIII/4x_foolhardy_Remacri/resolve/main/4x_foolhardy_Remacri.pth", "4x_foolhardy_Remacri.pth"),
    ("https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.2.4/RealESRGAN_x4plus_anime_6B.pth", "RealESRGAN_x4plus_anime_6B.pth"),
    #Juggernaut default Upscaler
    #("https://huggingface.co/uwg/upscaler/resolve/main/ESRGAN/4x_NMKD-Siax_200k.pth", "4x_NMKD-Siax_200k.pth"),
]
# Список URL VAE файлов.
VAE_URLS = [
    #Pony default VAE
    ("https://civitai.com/api/download/models/290640?type=VAE&format=SafeTensor", "sdxl_pony_vae.safetensors"),
]
# Список URL ControlNet моделей.
CONTROLNET_URLS = [
    #t2i-adapter_xl_openpose 151 MB
    #"https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/t2i-adapter_xl_openpose.safetensors",
    #t2i-adapter_xl_canny 148 MB
    #"https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/t2i-adapter_xl_canny.safetensors",
    #t2i-adapter_xl_sketch 148 MB
    #"https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/t2i-adapter_xl_sketch.safetensors",
    #t2i-adapter_diffusers_xl_depth_midas 151 MB
    #"https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/t2i-adapter_diffusers_xl_depth_midas.safetensors",
    #t2i-adapter_diffusers_xl_depth_zoe 151 MB
    #"https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/t2i-adapter_diffusers_xl_depth_zoe.safetensors",
    #t2i-adapter_diffusers_xl_lineart 151 MB
    #"https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/t2i-adapter_diffusers_xl_lineart.safetensors",
    #diffusers_xl_canny_full 2,5 GB"
    "https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/diffusers_xl_canny_full.safetensors",
    #diffusers_xl_depth_full 2,5 GB
    "https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/diffusers_xl_depth_full.safetensors",
    #thibaud_xl_openpose 2,5 GB"
     "https://huggingface.co/lllyasviel/sd_control_collection/resolve/main/thibaud_xl_openpose.safetensors",
    #xinsir_union_sdxl 2,5 GB
    #"https://huggingface.co/xinsir/controlnet-union-sdxl-1.0/resolve/main/diffusion_pytorch_model_promax.safetensors",
    #xinsir_openpose_sdxl 2,5 GB
    #"https://huggingface.co/xinsir/controlnet-openpose-sdxl-1.0/resolve/main/diffusion_pytorch_model.safetensors",
]
# Список URL IPAdapter моделей.
IPADAPTER_URLS = [
    ("https://huggingface.co/h94/IP-Adapter-FaceID/resolve/main/ip-adapter-faceid-plusv2_sdxl.bin", "ip-adapter-faceid-plusv2_sdxl.bin"),
]
#список клип
CLIP_VISION_URLS = [
    ("https://huggingface.co/h94/IP-Adapter/resolve/main/models/image_encoder/model.safetensors", "CLIP-ViT-H-14-laion2B-s32B-b79K.safetensors"),
]
#список ONNX моделей
BBOX_URLS = [
    ("https://huggingface.co/Ultralytics/YOLOv8/resolve/main/yolov8l.pt", "yolov8l.pt"),
    ("https://huggingface.co/Ultralytics/YOLOv8/resolve/main/yolov8m.pt", "yolov8m-face.pt"),
    ("https://huggingface.co/Ultralytics/YOLOv8/resolve/main/yolov8s.pt", "yolov8s-face.pt"),
    ("https://github.com/akanametov/yolov9-face/releases/download/1.0/yolov9-c-face.pt", "yolov9c-face.pt"),
]

SEGM_URLS = [
    ("https://huggingface.co/thejagstudio/textilev1Plain/resolve/main/yolov8m-seg.pt", "yolov8m-seg.pt"),
]

SAMS_URLS = [
    ("https://huggingface.co/scenario-labs/sam_vit/resolve/main/sam_vit_b_01ec64.pth", "sam_vit_b_01ec64.pth"),
    ("https://huggingface.co/scenario-labs/sam_vit/resolve/main/sam_vit_h_4b8939.pth", "sam_vit_h_4b8939.pth"),
]
def now_ts():
    """Текущее время для логов."""
    return time.strftime("%Y-%m-%d %H:%M:%S")

def log(msg):
    """Единый формат логов с таймстампом."""
    print(f"[{now_ts()}] {msg}")

def run_cmd(cmd, check=True, capture_output=False):
    """Функция запускает команду, печатает её и (опционально) возвращает вывод."""
    shell_cmd = " ".join(str(c) for c in cmd)
    log(f"[CMD] {shell_cmd}")
    started = time.time()
    result = subprocess.run(cmd, check=check, text=True, capture_output=capture_output)
    log(f"[RC={result.returncode}] {shell_cmd} ({time.time()-started:.1f}s)")
    if capture_output and result.stdout:
        stdout_preview = result.stdout[-1200:]
        if stdout_preview.strip():
            log(f"[STDOUT]\n{stdout_preview}")
    if capture_output and result.stderr:
        stderr_preview = result.stderr[-1200:]
        if stderr_preview.strip():
            log(f"[STDERR]\n{stderr_preview}")
    return result

def with_token_if_needed(url):
    """Добавляет токен в URL для HF/Civitai, если токен есть и он ещё не добавлен."""
    if "huggingface.co" in url and HF_TOKEN and "token=" not in url:
        separator = "&" if "?" in url else "?"
        return f"{url}{separator}token={HF_TOKEN}"
    if "civitai.com" in url and CIVITAI_TOKEN and "token=" not in url:
        separator = "&" if "?" in url else "?"
        return f"{url}{separator}token={CIVITAI_TOKEN}"
    return url

def file_name_from_url(url):
    """Извлекает имя файла из URL-адреса."""
    return url.split("?")[0].rstrip("/").split("/")[-1]

def parse_download_item(item):
    """Нормализует элемент списка загрузки в пару (url, custom_file_name)."""
    if isinstance(item, str):
        return item, ""
    if isinstance(item, (tuple, list)) and len(item) >= 1:
        url = item[0]
        custom_name = item[1] if len(item) > 1 else ""
        return str(url), str(custom_name or "")
    raise ValueError(f"Неподдерживаемый формат элемента загрузки: {item!r}")

def install_repo_requirements(repo_dir):
    req_file = Path(repo_dir) / "requirements.txt"
    if req_file.exists():
        run_cmd([sys.executable, "-m", "pip", "install", "-U", "-r", str(req_file)], check=False)

def ensure_custom_node(repo_url):
    repo_name = repo_url.rstrip("/").split("/")[-1]
    repo_dir = Path(CUSTOM_NODES_DIR) / repo_name
    if (repo_dir / ".git").exists():
        run_cmd(["git", "-C", str(repo_dir), "pull", "--ff-only"], check=False)
    else:
        run_cmd(["git", "clone", repo_url, str(repo_dir)], check=False)
    install_repo_requirements(repo_dir)

def is_probably_text_error_file(file_path):
    """Проверяет, похож ли файл на HTML/JSON с ошибкой вместо модели."""
    file_path = Path(file_path)
    if not file_path.exists() or file_path.stat().st_size > INVALID_TEXT_BYTES_LIMIT:
        return False
    with open(file_path, "rb") as fh:
        raw = fh.read(16384)
    sample = raw.decode("utf-8", errors="ignore").lower()
    markers = [
        "<html", "<!doctype html", "access denied", "unauthorized", "forbidden",
        "cloudflare", '"error"', "invalid token", "expired token", "captcha"
    ]
    return any(marker in sample for marker in markers)

def try_get_expected_size_bytes(url):
    """Получает Content-Length (если сервер его отдаёт), иначе возвращает 0."""
    result = run_cmd([
        "curl", "-sSIL", "--retry", "2", "--retry-delay", "2", "--connect-timeout", "25", "-A", "Mozilla/5.0", url,
    ], check=False, capture_output=True)
    if result.returncode != 0:
        return 0
    headers = str(result.stdout or "")
    for line in reversed(headers.splitlines()):
        low = line.lower()
        if low.startswith("content-type:"):
            log(f"[HEAD] content-type: {line.split(':',1)[1].strip()}")
        if low.startswith("content-length:"):
            value = line.split(":", 1)[1].strip()
            if value.isdigit():
                return int(value)
    return 0

def is_valid_download(file_path, expected_size=0):
    """Проверяет файл на размер, сигнатуру текстовой ошибки и (опционально) длину."""
    file_path = Path(file_path)
    if not file_path.exists():
        return False
    file_size = file_path.stat().st_size
    if file_size < MIN_VALID_FILE_BYTES:
        log(f"[BAD] слишком маленький файл: {file_path.name}={file_size} bytes")
        return False
    if is_probably_text_error_file(file_path):
        log(f"[BAD] файл похож на HTML/ошибку: {file_path.name}")
        return False
    if expected_size > 0:
        lower_bound = int(expected_size * 0.95)
        upper_bound = int(expected_size * 1.05)
        if not (lower_bound <= file_size <= upper_bound):
            log(f"[WARN] Размер {file_path.name}={file_size} != ожидаемому {expected_size}")
            return False
    return True

def download_with_aria2c(url, target_dir, file_name):
    """Скачивает файл через aria2c в щадящем режиме (меньше split/коннектов)."""
    result = run_cmd([
        "aria2c", url,
        "--follow-torrent=false",
        "--max-connection-per-server=2",
        "--split=2",
        "--min-split-size=10M",
        "--allow-overwrite=true",
        "--auto-file-renaming=false",
        "--file-allocation=none",
        "--retry-wait=5",
        "--max-tries=3",
        "--timeout=120",
        "--summary-interval=5",
        "--console-log-level=notice",
        "-d", str(target_dir),
        "-o", file_name,
    ], check=False)
    return result.returncode == 0

def download_with_curl(url, output_path):
    """Скачивает файл через curl с возобновлением и периодическим логом прогресса."""
    cmd = [
        "curl", "-fL", "--retry", "8", "--retry-delay", "3", "--retry-all-errors",
        "--connect-timeout", "30", "--max-time", str(CURL_TIMEOUT_SEC),
        "-A", "Mozilla/5.0",
        "-C", "-",
        "-o", str(output_path), url,
    ]
    log(f"[CURL] старт загрузки: {output_path.name}")
    started = time.time()
    proc = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    last_size = output_path.stat().st_size if output_path.exists() else 0
    while proc.poll() is None:
        time.sleep(10)
        current_size = output_path.stat().st_size if output_path.exists() else 0
        delta = max(0, current_size - last_size)
        speed = delta / 10.0
        elapsed = max(1.0, time.time() - started)
        avg_speed = current_size / elapsed
        log(f"[PROGRESS] {output_path.name}: {current_size/1024/1024:.2f} MB, +{delta/1024/1024:.2f} MB/10s, now {speed/1024/1024:.2f} MB/s, avg {avg_speed/1024/1024:.2f} MB/s")
        last_size = current_size
    output = proc.communicate()[0] or ""
    if output.strip():
        log(f"[CURL-OUT]\n{output[-1200:]}")
    ok = proc.returncode == 0
    log(f"[CURL] завершен rc={proc.returncode} за {time.time()-started:.1f}s")
    return ok

def download_one(url, target_dir, custom_file_name=""):
    """Скачивает один файл во временный .part и публикует только валидный результат."""
    target_dir = Path(target_dir)
    target_dir.mkdir(parents=True, exist_ok=True)
    safe_url = with_token_if_needed(url)
    file_name = custom_file_name.strip() if custom_file_name else file_name_from_url(url)
    output_path = target_dir / file_name
    tmp_path = target_dir / f"{file_name}.part"
    expected_size = try_get_expected_size_bytes(safe_url)

    if output_path.exists() and not FORCE_REDOWNLOAD:
        if is_valid_download(output_path, expected_size=expected_size):
            log(f"[SKIP] {output_path} уже существует ({output_path.stat().st_size} bytes)")
            return (url, "skip", str(output_path))
        log(f"[RETRY] {output_path} невалиден, перекачиваю")
        output_path.unlink(missing_ok=True)

    if FORCE_REDOWNLOAD and output_path.exists():
        log(f"[RETRY] FORCE_REDOWNLOAD=1, удаляю {output_path}")
        output_path.unlink(missing_ok=True)

    prefer_aria = ARIA2C_AVAILABLE and "civitai.com" not in safe_url

    for attempt in range(1, DOWNLOAD_RETRIES + 1):
        tmp_path.unlink(missing_ok=True)
        log(f"[TRY {attempt}/{DOWNLOAD_RETRIES}] {url}")

        downloaded = False
        if prefer_aria:
            downloaded = download_with_aria2c(safe_url, target_dir, tmp_path.name)
            if not downloaded:
                log("[WARN] aria2c завершился с ошибкой, пробую curl")
        if not downloaded:
            downloaded = download_with_curl(safe_url, tmp_path)

        if downloaded and is_valid_download(tmp_path, expected_size=expected_size):
            tmp_path.replace(output_path)
            log(f"[OK] {output_path} ({output_path.stat().st_size} bytes)")
            return (url, "ok", str(output_path))

        tmp_path.unlink(missing_ok=True)
        log(f"[WARN] Попытка {attempt} неудачна для {url}")
        if attempt < DOWNLOAD_RETRIES:
            time.sleep(DOWNLOAD_RETRY_DELAY_SEC)

    log(f"[FAIL] {url} -> {output_path} (файл отсутствует или похож на ошибку)")
    return (url, "fail", str(output_path))

for repo in CUSTOM_NODE_REPOS:
    ensure_custom_node(repo)

download_jobs = []
for item in CHECKPOINT_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, CHECKPOINTS_DIR, 1))
for item in CONTROLNET_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, CONTROLNET_DIR, 1))
for item in IPADAPTER_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, IPADAPTER_DIR, 1))
for item in LORA_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, LORA_DIR, 2))
for item in VAE_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, VAE_DIR, 3))
for item in UPSCALER_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, UPSCALER_DIR, 4))
for item in CLIP_VISION_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, CLIP_VISION_DIR, 1))
for item in BBOX_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, BBOX_DIR, 4))
for item in SEGM_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, SEGM_DIR, 4))
for item in SAMS_URLS:
    url, custom_file_name = parse_download_item(item)
    download_jobs.append((url, custom_file_name, SAMS_DIR, 2))

download_jobs = sorted(download_jobs, key=lambda item: item[3])
results = []

log(f"Параллельных загрузок: {MAX_PARALLEL_DOWNLOADS}")
with ThreadPoolExecutor(max_workers=MAX_PARALLEL_DOWNLOADS) as pool:
    future_map = {pool.submit(download_one, job[0], job[2], job[1]): job for job in download_jobs}
    for future in as_completed(future_map):
        results.append(future.result())

ok_count = sum(1 for _, status, _ in results if status == "ok")
skip_count = sum(1 for _, status, _ in results if status == "skip")
fail_count = sum(1 for _, status, _ in results if status == "fail")
log(f"Загрузки завершены: ok={ok_count}, skip={skip_count}, fail={fail_count}")
if fail_count:
    log("Есть неуспешные загрузки, проверьте ссылки/токены. При необходимости включите FORCE_REDOWNLOAD=1")
```

```python
# ЯЧЕЙКА 5: запуск ComfyUI в двух режимах по выбору пользователя.
import base64  # Импорт base64 нужен для формирования заголовка Basic Auth в локальном proxy.
import os  # Импорт os нужен для переменных окружения.
import select  # Импорт select нужен для неблокирующего чтения stdout процесса.
import subprocess  # Импорт subprocess нужен для запуска процессов ComfyUI/cloudflared.
import textwrap  # Импорт textwrap нужен для записи скрипта локального auth-proxy.
import time  # Импорт time нужен для задержек при ожидании URL туннеля.
import urllib.request  # Импорт urllib нужен для health-check HTTP порта ComfyUI.
from getpass import getpass  # Импорт getpass нужен для безопасного ввода tunnel token без отображения.
from pathlib import Path  # Импорт Path нужен для проверки существования main.py.
import sys  # Импорт sys нужен для обращения к текущему интерпретатору Python (sys.executable).


comfy_main = Path(COMFY_DIR) / "main.py"
if not comfy_main.exists():
    raise FileNotFoundError(f"Не найден файл запуска ComfyUI: {comfy_main}")


def wait_comfy_ready(comfy_proc, timeout_sec=90):
    """Ждём доступности ComfyUI и параллельно печатаем логи старта."""
    start = time.time()
    health_url = "http://127.0.0.1:18188/system_stats"
    while time.time() - start < timeout_sec:
        if comfy_proc.poll() is not None:
            print(f"[ComfyUI] Процесс завершился с кодом {comfy_proc.returncode}")
            return False
        ready_to_read, _, _ = select.select([comfy_proc.stdout], [], [], 0.5)
        if ready_to_read:
            line = comfy_proc.stdout.readline()
            if line:
                print(f"[ComfyUI] {line.rstrip()}")
        try:
            with urllib.request.urlopen(health_url, timeout=2) as response:
                if response.status == 200:
                    print("[ComfyUI] Health-check OK: UI запущен и отвечает на /system_stats")
                    return True
        except Exception:
            pass
    print("[ComfyUI] Таймаут ожидания запуска. Проверьте логи выше.")
    return False


def ensure_auth_proxy_script(proxy_script_path):
    """Создаём локальный reverse proxy с Basic Auth (поддержка websocket)."""
    script_text = textwrap.dedent("""
        import argparse
        import asyncio
        import base64
        from aiohttp import web, ClientSession, WSMsgType

        parser = argparse.ArgumentParser()
        parser.add_argument("--listen-host", default="127.0.0.1")
        parser.add_argument("--listen-port", type=int, default=8787)
        parser.add_argument("--upstream", default="http://127.0.0.1:18188")
        parser.add_argument("--auth", required=True)
        args = parser.parse_args()

        expected = "Basic " + base64.b64encode(args.auth.encode("utf-8")).decode("utf-8")

        async def handle_request(request):
            auth_header = request.headers.get("Authorization", "")
            if auth_header != expected:
                return web.Response(status=401, headers={"WWW-Authenticate": 'Basic realm="ComfyUI"'}, text="Authentication required")

            upstream_url = f"{args.upstream}{request.rel_url}"
            conn_hdr = request.headers.get("Connection", "")
            is_ws = "upgrade" in conn_hdr.lower() and request.headers.get("Upgrade", "").lower() == "websocket"

            async with ClientSession() as session:
                if is_ws:
                    ws_server = web.WebSocketResponse()
                    await ws_server.prepare(request)
                    async with session.ws_connect(upstream_url, headers={k: v for k, v in request.headers.items() if k.lower() != "host"}) as ws_client:
                        async def client_to_upstream():
                            async for msg in ws_server:
                                if msg.type == WSMsgType.TEXT:
                                    await ws_client.send_str(msg.data)
                                elif msg.type == WSMsgType.BINARY:
                                    await ws_client.send_bytes(msg.data)
                                elif msg.type == WSMsgType.CLOSE:
                                    await ws_client.close()

                        async def upstream_to_client():
                            async for msg in ws_client:
                                if msg.type == WSMsgType.TEXT:
                                    await ws_server.send_str(msg.data)
                                elif msg.type == WSMsgType.BINARY:
                                    await ws_server.send_bytes(msg.data)
                                elif msg.type in (WSMsgType.CLOSE, WSMsgType.CLOSED, WSMsgType.ERROR):
                                    await ws_server.close()
                                    break

                        await asyncio.gather(client_to_upstream(), upstream_to_client())
                    return ws_server

                body = await request.read()
                headers = {k: v for k, v in request.headers.items() if k.lower() not in {"host", "content-length", "authorization"}}
                async with session.request(request.method, upstream_url, headers=headers, data=body, allow_redirects=False) as resp:
                    resp_body = await resp.read()
                    response_headers = {k: v for k, v in resp.headers.items() if k.lower() != "content-length"}
                    return web.Response(status=resp.status, headers=response_headers, body=resp_body)

        app = web.Application(client_max_size=1024**3)
        app.router.add_route('*', '/{path_info:.*}', handle_request)
        web.run_app(app, host=args.listen_host, port=args.listen_port)
    """).strip() + "\n"
    proxy_script_path.write_text(script_text, encoding="utf-8")


print("1) вариант запуска: локальный режим без доступа извне (для SSH-туннеля на Vast.ai)")
print("2) вариант запуска: полноценный Cloudflare Tunnel через token + обязательный GRADIO_AUTH=login:password")
launch_mode = input("Введите режим запуска (1/2): ").strip()
if launch_mode not in {"1", "2"}:
    raise ValueError("Нужно ввести только 1 или 2")

comfy_cmd = [sys.executable, str(comfy_main), "--port", "18188", "--listen", "127.0.0.1"]

if launch_mode == "1":
    # проверяем GRADIO_AUTH
    access_auth = os.environ.get("GRADIO_AUTH", "").strip()
    if not access_auth or ":" not in access_auth:
        raise ValueError("Для локального режима через SSH обязательно задайте GRADIO_AUTH в формате login:password")

    print("Запуск локального режима ComfyUI. Используйте SSH-туннель к 127.0.0.1:8787")
    comfy_proc = subprocess.Popen(comfy_cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, bufsize=1)
    is_ready = wait_comfy_ready(comfy_proc)

    # создаём и запускаем локальный proxy
    proxy_script = BASE_DIR / "comfy_auth_proxy.py"
    ensure_auth_proxy_script(proxy_script)
    proxy_proc = subprocess.Popen(
        [sys.executable, str(proxy_script),
         "--listen-host", "127.0.0.1",
         "--listen-port", "8787",
         "--upstream", "http://127.0.0.1:18188",
         "--auth", access_auth],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1,
    )

    print("Локальный auth-proxy поднят. Через SSH-туннель заходите на http://127.0.0.1:8787/ и вводите ваш GRADIO_AUTH")
else:
    access_auth = os.environ.get("GRADIO_AUTH", "").strip()
    if not access_auth or ":" not in access_auth:
        raise ValueError("Для публичного режима обязательно задайте GRADIO_AUTH в формате login:password")

    tunnel_token = os.environ.get("CLOUDFLARE_TUNNEL_TOKEN", "").strip()
    if not tunnel_token:
        tunnel_token = getpass("Введите CLOUDFLARE_TUNNEL_TOKEN (ввод скрыт): ").strip()
    if not tunnel_token:
        raise ValueError("Не передан CLOUDFLARE_TUNNEL_TOKEN. Создайте tunnel token в Cloudflare Zero Trust и запустите снова.")

    if subprocess.run([sys.executable, "-c", "import aiohttp"], check=False).returncode != 0:
        print("aiohttp не найден, устанавливаю...")
        subprocess.run([sys.executable, "-m", "pip", "install", "-q", "aiohttp"], check=False)

    print("Запуск ComfyUI для внешнего доступа через auth-proxy...")
    comfy_proc = subprocess.Popen(comfy_cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, bufsize=1)
    is_ready = wait_comfy_ready(comfy_proc)

    proxy_script = BASE_DIR / "comfy_auth_proxy.py"
    ensure_auth_proxy_script(proxy_script)
    proxy_proc = subprocess.Popen(
        [sys.executable, str(proxy_script), "--listen-host", "127.0.0.1", "--listen-port", "8787", "--upstream", "http://127.0.0.1:8188", "--auth", access_auth],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1,
    )
    time.sleep(2)
    if proxy_proc.poll() is not None:
        print("[AUTH_PROXY] Процесс завершился раньше времени. Логи:")
        print(proxy_proc.stdout.read())
        raise RuntimeError("Не удалось запустить auth proxy")

    cloudflared_exists = subprocess.run(["bash", "-lc", "command -v cloudflared >/dev/null"], check=False).returncode == 0
    if not cloudflared_exists:
        subprocess.run([
            "bash", "-lc",
            "wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O /usr/local/bin/cloudflared && chmod +x /usr/local/bin/cloudflared"
        ], check=False)

    tunnel_cmd = ["cloudflared", "tunnel", "--no-autoupdate", "run", "--token", tunnel_token, "--url", "http://127.0.0.1:8787"]
    print("Запуск Cloudflare Tunnel по token...")
    print("Команда:", " ".join(["cloudflared", "tunnel", "--no-autoupdate", "run", "--token", "***", "--url", "http://127.0.0.1:8787"]))
    tunnel_proc = subprocess.Popen(
        tunnel_cmd,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1,
    )

    print("Cloudflare Tunnel запущен. Публичный URL берётся из настроек вашего named tunnel в Cloudflare.")
    print("Если tunnel не стартует, проверьте логи ниже (обычно проблема в token или маршруте hostname -> service).")
    start_time = time.time()
    while time.time() - start_time < 45:
        line = tunnel_proc.stdout.readline()
        if not line:
            continue
        print(f"[cloudflared] {line.rstrip()}")
        if "Registered tunnel connection" in line or "Connection" in line:
            break

    auth_header = "Basic " + base64.b64encode(access_auth.encode("utf-8")).decode("utf-8")
    print("Туннель поднят через token. Откройте ваш домен, назначенный этому tunnel, в панели Cloudflare Zero Trust.")
    print("Проверка auth заголовка (для curl, подставьте ваш домен):")
    print(f"curl -I -H 'Authorization: {auth_header}' https://YOUR_TUNNEL_HOSTNAME")
    if not is_ready:
        print("Внимание: tunnel поднят, но health-check ComfyUI не подтвердил запуск.")
```
